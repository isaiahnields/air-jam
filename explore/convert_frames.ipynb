```
import cv2
import numpy as np
import os

vidcap = cv2.VideoCapture('VID_20190330_112856.mp4')
success,image = vidcap.read()
count = 0

while success:
  cv2.imwrite("frame_drums_isi_%d.jpg" % count, image)     # save frame as JPEG file      
  success,image = vidcap.read()
  print('Read a new frame: ', success)
  count += 1
```
